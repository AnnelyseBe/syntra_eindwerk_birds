## Input species id for transformation

```python

species_id = 70
```

```python
species = { 
        116 : {'afkorting' : 'hp', 'soort' : 'halsbandparkiet'}, 
        70 : {'afkorting' : 'bk', 'soort' : 'boomklever'}
        }
```

## Load clean data

```python
# Load the data
observations = f'../2_cleaning/clean_data/observations_{species_id}_clean.parquet'
df_observations = pd.read_parquet(observations, engine="pyarrow")
```

## Transform data

```python
df_observations.reset_index(drop=False, inplace=True)
```

```python
df_observations_gold = df_observations[['species_id', 'species_name', 'date', 'observer_id', 'latitude', 'longitude', 'location_id', 'observation_id']]
```

## Write result to parquet-file in "gold" folder

```python
df_observations_gold.to_parquet(f'./gold/observations_{species[species_id]['afkorting']}.parquet', index= True, engine="pyarrow")
```
