```python
df_first_cleaning = df.dropna(subset=['species_id'])
```

## Species_id
De species_id is niet degene die we verwachten. Bij de general observations hebben we op een specifieke soort gezocht, maar nu blijkt dat bij de observation_details de boomklever, dat deze niet altijd overeenkomt met degene die we wilden ophalen. </br>

#### conclusie
Bij de boomklevers zien we dat we ook een species_id 258977 en 258978 hebben. Dit blijken ondersoorten van de boomklever te zijn , nl Boomklever ssp caesia en Boomklever ssp europaea </br>
In onze oefening gaan we het onderscheid tussen deze soorten niet maken, deze ondersoort blijft behouden in de data.

```python
df_first_cleaning.groupby('species_id').nunique()
```

```python
df1 = df_first_cleaning[df_first_cleaning['species_id'] == 70].head(2)
df2 = df_first_cleaning[df_first_cleaning['species_id'] == 258977.0].head(2)
df3 = df_first_cleaning[df_first_cleaning['species_id'] == 258978.0].head(2)

pd.concat([df1, df2, df3], axis=0)
```
