# Samenvoegen van de verschillende ophalingen per soort

Vermits het ophalen van de general observaties in verschillende sessies is opgehaald, moeten de verschillende .csv files gecombineerd worden. </br>
Duplicates in deze gecombineerde dataset, door dubbel ophalen van data, worden verwijderd. 
Clean data wordt (per soort nl. halsbandparkieten en boomklevers) in een aparte map gezet.



```python
import pandas as pd
import os
from random import randint


###########################################################################################################################################
species_id = 70
folder_path = "./scraped_data/"
folder_path_clean = "./scraped_data/cleaned/"
temp_combined_csv = "temp_combined_observations_delete_me.csv"

# Find all CSV files containing "observations_{species_id}" in their filename
csv_files = [f for f in os.listdir(folder_path) if f"observations_{species_id}" in f and f.endswith(".csv")]
# Sort the files, prioritizing those containing "pages_1- because the header needs to be put on top
csv_files.sort(key=lambda x: 0 if "pages_1-" in x else 1)

# Append all CSV files into a single new CSV file
with open(temp_combined_csv, "w") as outfile:
    for i, file in enumerate(csv_files):
        file_path = os.path.join(folder_path, file)
        with open(file_path, "r") as infile:
                outfile.write(infile.read())

# Load the appended CSV file into a DataFrame
combined_df = pd.read_csv(temp_combined_csv)

# Remove duplicates
cleaned_df = combined_df.drop_duplicates()

# Save the cleaned DataFrame to a new CSV file
cleaned_df.to_csv(os.path.join(folder_path_clean, f"observations_general_{species_id}_clean.csv"), index=False)

# Optional: Remove the temporary combined file
os.remove(temp_combined_csv)          
```
